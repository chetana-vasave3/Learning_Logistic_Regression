# Using mannual formula


```python
tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()

true_positive = tp
false_positive = fp
false_negative = fn
true_negative = tn
```

```python
# Accuracy

Accuracy = (true_positive + true_negative) / (true_positive +false_positive + false_negative + true_negative)
print("Accuracy",Accuracy)

# Recall

# Recall
Recall = true_positive/(true_positive+false_negative)
print("Recall",Recall)

# Precison

# Precison
Precision = true_positive/(true_positive+false_positive)
print("Precision",Precision)

# F1 Score

# F1 Score
F1_Score = 2*(Recall * Precision) / (Recall + Precision)
print("F1_Score",F1_Score)
```

```python
def classification_evalution(x,y,model):
    
  
    
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LogisticRegression,Lasso,Ridge
    from sklearn import metrics
    import matplotlib.pyplot as plt
    import warnings 
    warnings.filterwarnings("ignore")
    import pandas as pd
    import seaborn as sns
    import numpy as np
    
    #splitting into test train
    test_data_percentage=float(input("Enter a percentage value you want to split your test data : "))
    x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=test_data_percentage,random_state=355)
    
    #model Fitting
    model.fit(x_train,y_train)
    
    #Calculate y predicted for testing
    y_pred_test=model.predict(x_test)
    
    #Evalution
    
    #Plotting Confusion matrix
    print("Confusion matrix is ")
    from sklearn import metrics
    metrics.plot_confusion_matrix(model,x_test, y_test)
    plt.show()
    
    tn, fp, fn, tp = confusion_matrix(y_test,y_pred_test).ravel()

    true_positive = tp
    false_positive = fp
    false_negative = fn
    true_negative = tn
 

    # Breaking down the formula for Accuracy
    Accuracy = (true_positive + true_negative) / (true_positive +false_positive + false_negative + true_negative)

    # Precison
    Precision = true_positive/(true_positive+false_positive)

    # Recall
    Recall = true_positive/(true_positive+false_negative)
   
    # F1 Score
    F1_Score = 2*(Recall * Precision) / (Recall + Precision)
    
    
    #Calculation of diff matrices
    fpr,tpr,threhsolds=metrics.roc_curve(y_test,y_pred_test)
    
    auc=metrics.roc_auc_score(y_test,y_pred_test)
    

    print("Classification Report is \n",metrics.classification_report(y_test,y_pred_test))
    
    #ROC plotting
    plt.plot(fpr, tpr, color='orange', label='ROC curve (area = %0.3f)' % auc)
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--',label='ROC base model')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.show()
    
    #taking values in Dataframe
    values=np.array([[Accuracy],[Precision],[Recall],[F1_Score],[auc]])
    index1=["Accuracy","Precision","Recall","F1_Score","AUC "]
    
    ans=pd.DataFrame(data=values,index=index1,columns=["Values"],dtype="float")  
    pd.set_option('display.float_format', '{:.4f}'.format)
    
    return ans


classification_evalution(X,y,log_reg)
```
